## The WindModule
#### Designed to work with wind data for a set of wind turbines. It provides functionalities to download wind data, calculate mean wind speed in a specific direction, visualize wind data in the form of wind roses, and calculate some statistics.

#### __init__(self, wind_turbines, start_year=2016, end_year=2018):
##### Initializes the attributes of the class and calls the download_data method to download the data for the given turbines and years.
##### download_data(self):
This method downloads wind data for each turbine from the New European Wind Atlas API.
For each turbine, it constructs a URL using the turbine's position and the specified date range.
It then sends a GET request to the constructed URL.
If the request is successful, it saves the response content to a temporary .nc (NetCDF) file and then converts this file to a pandas DataFrame.
The DataFrame is then stored in the dataframes dictionary with the turbine's name as the key.

```python
wind_module = WindModule(wind_turbines,2016,2017) #if no specific dates are given, downloads data from 2016 to 2018 
```

## Let's go through the noisemap class

### This class, NoiseMap, is designed for generating and manipulating noise maps, focusing on the noise produced by multiple wind turbines.

#### Initializes the class with specified parameters, including a list of wind turbines, a wind turbine model, an air absorption coefficient, and a wind speed.
#### Key methods:
##### calculate_sound_intensity_level: Computes the sound intensity level at a specified distance from the source, considering air absorption.
##### convert_intensity_level_into_dB: Converts a given sound intensity level into decibels (dB).
##### superpose_several_wind_turbine_sounds: Superposes sounds from multiple wind turbines at specific observation points and returns the total noise levels in dB at those points.
##### generate_noise_map: Generates a noise map for the wind turbines and observation points, storing the values in the object's attributes.
##### plot_noise_map: Plots the generated noise map with wind turbines and observation points marked.
##### display_turbines_on_map: Displays the wind turbines and observation points on a real map using folium.
##### calculate_noise_at_position: Calculates the noise level at a given observation point.
##### generate_and_plot_noise_map: Generates and plots a noise map for the wind turbines and observation points based on a specified desired noise level.
##### get_landuse_between_points: Retrieves land use information between two specified points.
##### plot_landuse_for_turbines_and_observation_points: Plots land use, turbines, and observation points on a map and draws lines between turbine positions and observation points.
##### get_altitudes_between_points: Retrieves altitudes between two specified points using the Google Elevation API.
##### plot_relief_between_points: Plots the relief (altitude variations) between each wind turbine and each observation point.
